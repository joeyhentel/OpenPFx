```python
import pandas as pd
sfew_df = pd.read_csv('pfx_single_few.csv')
```

```python
mfew_df = pd.read_csv('pfx_mf.csv')
```

```python
finding_order = {finding : i for i, finding in enumerate(sfew_df['finding'].unique())}
```

```python
mfew_df['sort_order'] = mfew_df['finding'].map(finding_order)
```

```python
mfew_df = mfew_df.sort_values('sort_order', kind = 'stable').drop(columns = ['sort_order']).reset_index(drop = True)
```
