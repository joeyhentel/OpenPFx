```python
import pandas as pd

# Load CSV
df = pd.read_csv("mf_2.csv")

# For every 5 rows, pick the row with the highest score
result = df.loc[df.groupby(df.index // 5)["relaxed_overall_score"].idxmax()]

# Reset index for cleanliness
result = result.reset_index(drop=True)

# Show result
display(result)

# Save to new CSV if needed
result.to_csv("mf_cleaned_2.csv", index=False)
```
